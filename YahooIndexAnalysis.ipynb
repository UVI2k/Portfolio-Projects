### We choose dataset of "Master Card", containing the stock price history

```python
mastercard_df.sort_values(by = 'Date', ascending=True)
```

```python
mastercard_df['Date'] = pd.to_datetime(mastercard_df['Date'])
mastercard_df.set_index('Date', inplace = True)
```

```python
mastercard_df['Average'] = (mastercard_df['High'] + mastercard_df['Low'])/2
mastercard_df['Average']
```

```python
mastercard_df[['Open', 'High', 'Low', 'Close', 'Volume']].describe()
```

```python
mastercard_df.info()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (15, 8))
plt.plot(mastercard_df['Date'], mastercard_df['Open'])
plt.title('Stock Performance Over Time')
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.show()
```

```python
mastercard_df['Average_44D'] = mastercard_df['Average'].rolling(window = 44).mean()
mastercard_df.tail()
```

```python
plt.figure(figsize=(15,8))
plt.plot(mastercard_df.index, mastercard_df['Average'], color = 'blue')
plt.plot(mastercard_df.index, mastercard_df['Average_44D'], color = 'orange')
plt.title("Moving Average 44 days")
plt.xlabel("Date")
plt.ylabel("Stock Price and Moving Average")
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.bar(mastercard_df.index, mastercard_df['Volume'], color = 'purple', alpha = 0.5)
plt.title('Volume Analysis')
plt.xlabel('Date')
plt.ylabel("Volume")
plt.show()
```

```python
import seaborn as sns

corr  = mastercard_df[['Open', 'High', 'Low', 'Close', 'Volume']].corr()
sns.heatmap(corr, annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.title('Correlation Matrix of the Features')
plt.show
```
